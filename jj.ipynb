```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

url = "https://web.stanford.edu/class/archive/cs/cs109/cs109.1166/stuff/titanic.csv"
data = pd.read_csv(url)

print("Первые 5 строк данных:")
print(data.head())

print("\nОписание признаков:")
print("1. Survived - Выжил (0 = Нет, 1 = Да)")
print("2. Pclass - Класс билета (1 = 1-й, 2 = 2-й, 3 = 3-й)")
print("3. Name - Имя пассажира")
print("4. Sex - Пол")
print("5. Age - Возраст")
print("6. Siblings/Spouses Aboard - Количество братьев/сестер/супругов на борту")
print("7. Parents/Children Aboard - Количество родителей/детей на борту")
print("8. Fare - Стоимость билета")

print("\nФорма данных:", data.shape)
print("\nОбщая информация о данных:")
print(data.info())
print("\nПропущенные значения:")
print(data.isnull().sum())

print("\nОсновные статистические показатели:")
print(data.describe())

sns.set_style("whitegrid")
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
sns.histplot(data['Age'].dropna(), bins=30, kde=True)
plt.title('Распределение возраста')

plt.subplot(2, 2, 2)
sns.boxplot(x='Pclass', y='Age', data=data)
plt.title('Распределение возраста по классам')

plt.subplot(2, 2, 3)
sns.countplot(x='Survived', hue='Sex', data=data)
plt.title('Выживаемость по полу')

plt.subplot(2, 2, 4)
sns.heatmap(data.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Корреляционная матрица')

plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 5))
sns.histplot(data['Fare'], bins=40, kde=True, color='blue', label='Оригинал')
sns.histplot(-data['Fare'], bins=40, kde=True, color='red', label='Отражение')

plt.title('Распределение стоимости билетов (отражено)')
plt.xlabel('Стоимость билетов (Fare)')
plt.legend()
plt.show()

plt.figure(figsize=(8, 5))
sns.barplot(x='Pclass', y='Survived', data=data)
plt.title('Процент выживших по классам')
plt.ylabel('Вероятность выживания')
plt.show()
```
